### **The Answer for the Mid-Exam**

```python
from pulp import LpMaximize, LpProblem, LpVariable, LpInteger

# Create a linear programming problem
prob = LpProblem(name="Maximize_Profit", sense=LpMaximize)

# Define decision variables
# variabels for the product amounts
x1 = LpVariable(name="X1", cat=LpInteger)
x2 = LpVariable(name="X2", cat=LpInteger)
x3 = LpVariable(name="X3", cat=LpInteger)
x4 = LpVariable(name="X4", cat=LpInteger)

# variables for the choice of products
y1 = LpVariable(name="Y1", cat="Binary")
y2 = LpVariable(name="Y2", cat="Binary")
y3 = LpVariable(name="Y3", cat="Binary")
y4 = LpVariable(name="Y4", cat="Binary")

# variables for the choice of plant
z1 = LpVariable(name="Z1", cat="Binary")
z2 = LpVariable(name="Z2", cat="Binary")
z3 = LpVariable(name="Z3", cat="Binary")

# Define the objective function
prob += 5*x1 + 7*x2 + 3*x3 + 5*x4 - 8*y1 - 8*y2 - 5*y3 - 4*y4, "Profit"

# Add constraints
prob += y1 + y2 + y3 + y4 <= 3, "Constraint1" # choice of products <= 3
prob += z1 + z2 + z3 == 1, "Constraint2" # only one plant

# constraints for the production quantity of each product
prob += x1 <= 7*y1, "Constraint3"
prob += x2 <= 5*y2, "Constraint4"
prob += x3 <= 9*y3, "Constraint5"
prob += x4 <= 8*y4, "Constraint6"

# factory constraints
M = 1000
prob += 3*x1 + 4*x2 + 2*x3 + 0*x4 <= 30 + M*(1-z1), "Constraint7"
prob += 4*x1 + 6*x2 + 2*x3 + 3*x4 <= 40 + M*(1-z2), "Constraint8"
prob += 3*x1 + 5*x2 + 3*x3 + 2*x4 <= 35 + M*(1-z3), "Constraint9"

# non-negativity constraint
prob += x1 >= 0, "Constraint10"
prob += x2 >= 0, "Constraint11"
prob += x3 >= 0, "Constraint12"
prob += x4 >= 0, "Constraint13"

# Solve the problem
prob.solve()

# Print the results
print("Status:", prob.status)
print("Objective Value (Profit):", round(prob.objective.value(), 2))

# Print the values of decision variables
for var in prob.variables():
    print(f"{var.name} = {var.value()}")
```
